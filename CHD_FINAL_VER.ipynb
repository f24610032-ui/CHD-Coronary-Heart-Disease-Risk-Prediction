```python
# 2. IMPORTS
# ===============================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score, roc_curve, make_scorer, f1_score, precision_recall_curve
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.svm import SVC

import xgboost as xgb
!pip install catboost
from catboost import CatBoostClassifier
from imblearn.over_sampling import SMOTE
import tensorflow as tf

print("\n SHAPE (Rows, Columns) ")
# The 'df' DataFrame is not yet loaded at this point, so these print statements will cause an error.
# Moving them after the data loading section.


# ===============================
# 3. LOAD DATA
# ===============================
df = pd.read_csv("/content/framingham.csv")
df = df.dropna()

# Moving initial data prints here after df is loaded
print(df.shape)

print("\n---- DATA TYPES ----")
print(df.dtypes)

print("\n---- FIRST 5 ROWS ----")
print(df.head())

# ===============================
# 4. EDA
# ===============================
sns.countplot(x='TenYearCHD', data=df)
plt.title("10-Year CHD Cases Distribution")
plt.show()

sns.histplot(df['age'], kde=True)
plt.title("Distribution of Age")
plt.show()

sns.countplot(x='male', data=df)
plt.title("Male vs Female")
plt.show()

sns.countplot(x='currentSmoker', data=df)
plt.title("Smokers vs Non-Smokers")
plt.show()

sns.boxplot(x=df['totChol'])
plt.title("Total Cholesterol Outliers")
plt.show()

plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), cmap='coolwarm', annot=False)
plt.title("Correlation Heatmap")
plt.show()

# ===============================
# 5. FEATURE STATISTICS
# ===============================
numeric_stats = df.describe().T
numeric_stats['median'] = df.median()
numeric_stats['mode'] = df.mode().iloc[0]
numeric_stats['variance'] = df.var()
numeric_stats['skewness'] = df.skew()
numeric_stats['kurtosis'] = df.kurt()
print("\n=== Numeric Columns Statistics ===")
print(numeric_stats)

# ===============================
# 6. TRAIN-TEST SPLIT
# ===============================
X = df.drop('TenYearCHD', axis=1)
y = df['TenYearCHD']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
print("Before SMOTE:\n", y_train.value_counts())

# ===============================
# 7. APPLY SMOTE
# ===============================
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)
print("After SMOTE:\n", y_train_res.value_counts())

# ===============================
# 8. XGBOOST MODEL
# ===============================
xgb_model = xgb.XGBClassifier(
    n_estimators=200,
    max_depth=5,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    use_label_encoder=False,
    eval_metric='logloss'
)
xgb_model.fit(X_train_res, y_train_res)
y_pred_xgb = xgb_model.predict(X_test)
y_pred_proba_xgb = xgb_model.predict_proba(X_test)[:,1]

print("\n=== XGBoost Results ===")
print("Accuracy:", accuracy_score(y_test, y_pred_xgb))
print("AUC:", roc_auc_score(y_test, y_pred_proba_xgb))
print(confusion_matrix(y_test, y_pred_xgb))
print(classification_report(y_test, y_pred_xgb))

xgb.plot_importance(xgb_model, max_num_features=10)
plt.title("XGBoost Feature Importance")
plt.show()

fpr_xgb, tpr_xgb, _ = roc_curve(y_test, y_pred_proba_xgb)
plt.figure(figsize=(8,6))
plt.plot(fpr_xgb, tpr_xgb, label=f"XGBoost AUC = {roc_auc_score(y_test, y_pred_proba_xgb):.3f}")
plt.plot([0,1],[0,1],'--', color='gray')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - XGBoost")
plt.legend()
plt.show()

# ===============================
# 9. RANDOM FOREST MODEL
# ===============================
rf_model = RandomForestClassifier(
    n_estimators=300,
    max_depth=8,
    random_state=42,
    class_weight='balanced'
)
rf_model.fit(X_train_res, y_train_res)
y_pred_rf = rf_model.predict(X_test)
y_pred_proba_rf = rf_model.predict_proba(X_test)[:,1]

print("\n=== Random Forest Results ===")
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("AUC:", roc_auc_score(y_test, y_pred_proba_rf))
print(confusion_matrix(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))

fi_df_rf = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': rf_model.feature_importances_
}).sort_values(by='Importance', ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(data=fi_df_rf, x='Importance', y='Feature')
plt.title("Random Forest Feature Importance")
plt.show()

fpr_rf, tpr_rf, _ = roc_curve(y_test, y_pred_proba_rf)
plt.figure(figsize=(8,6))
plt.plot(fpr_rf, tpr_rf, label=f"Random Forest AUC = {roc_auc_score(y_test, y_pred_proba_rf):.3f}")
plt.plot([0,1],[0,1],'--', color='gray')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Random Forest")
plt.legend()
plt.show()

# ===============================
# 10. CATBOOST MODEL
# ===============================
categorical_cols = ['male', 'currentSmoker', 'BPMeds', 'prevalentStroke', 'prevalentHyp', 'diabetes']
for col in categorical_cols:
    X_train_res[col] = X_train_res[col].astype(int)
    X_test[col] = X_test[col].astype(int)

# Feature Engineering
X_train_res['age_sysBP'] = X_train_res['age'] * X_train_res['sysBP']
X_test['age_sysBP'] = X_test['age'] * X_test['sysBP']

X_train_res['chol_age_ratio'] = X_train_res['totChol'] / (X_train_res['age'] + 1)
X_test['chol_age_ratio'] = X_test['totChol'] / (X_test['age'] + 1)

X_train_res['BMI_sysBP'] = X_train_res['BMI'] * X_train_res['sysBP']
X_test['BMI_sysBP'] = X_test['BMI'] * X_test['sysBP']

print("Feature engineering completed:", ['age_sysBP', 'chol_age_ratio', 'BMI_sysBP'])

cat_model = CatBoostClassifier(
    iterations=500,
    learning_rate=0.05,
    depth=6,
    eval_metric='AUC',
    random_seed=42,
    verbose=0,
    class_weights=[1, 5]
)
cat_model.fit(X_train_res, y_train_res, cat_features=categorical_cols)
y_pred_cb = cat_model.predict(X_test)
y_pred_proba_cb = cat_model.predict_proba(X_test)[:,1]

print("\n=== CatBoost Results ===")
print("Accuracy:", accuracy_score(y_test, y_pred_cb))
print("AUC:", roc_auc_score(y_test, y_pred_proba_cb))
print(confusion_matrix(y_test, y_pred_cb))
print(classification_report(y_test, y_pred_cb))

# ===============================
# 11. LOGISTIC REGRESSION
# ===============================
param_grid = {
    'C':[0.01,0.1,1,10],
    'penalty':['l2'],
    'solver':['lbfgs'],
    'class_weight':['balanced'],
    'max_iter':[2000]
}
log_model = LogisticRegression(random_state=42)
grid = GridSearchCV(log_model, param_grid, scoring='roc_auc', cv=5)
grid.fit(X_train_res, y_train_res)
best_log_model = grid.best_estimator_
print("Best Logistic Hyperparameters:", grid.best_params_)

y_pred_proba = best_log_model.predict_proba(X_test)[:,1]
precisions, recalls, thresholds = precision_recall_curve(y_test, y_pred_proba)
f1_scores = 2*precisions*recalls/(precisions+recalls+1e-8)
best_thresh = thresholds[np.argmax(f1_scores)]
y_pred = (y_pred_proba >= best_thresh).astype(int)

print("\n=== Logistic Regression Results ===")
print("Accuracy:", accuracy_score(y_test, y_pred))
print("AUC:", roc_auc_score(y_test, y_pred_proba))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))

# ===============================
# 12. PCA (OPTIONAL)
# ===============================
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_res)
X_test_scaled = scaler.transform(X_test)

pca = PCA(n_components=0.95)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)
print("Original features:", X_train_res.shape[1], "-> PCA features:", X_train_pca.shape[1])

# ===============================
# 13. SVM
# ===============================
svm_model = SVC(
    kernel='rbf',
    C=1.0,
    gamma='scale',
    probability=True,
    class_weight='balanced',
    random_state=42
)
svm_model.fit(X_train_scaled, y_train_res)
y_pred_svm = svm_model.predict(X_test_scaled)
y_pred_proba_svm = svm_model.predict_proba(X_test_scaled)[:,1]

print("\n=== SVM Results ===")
print("Accuracy:", accuracy_score(y_test, y_pred_svm))
print("AUC:", roc_auc_score(y_test, y_pred_proba_svm))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_svm))
print("\nClassification Report:\n", classification_report(y_test, y_pred_svm))

fpr_svm, tpr_svm, _ = roc_curve(y_test, y_pred_proba_svm)
plt.figure(figsize=(8,6))
plt.plot(fpr_svm, tpr_svm, label=f"SVM AUC = {roc_auc_score(y_test, y_pred_proba_svm):.3f}")
plt.plot([0,1],[0,1],'--', color='gray')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - SVM")
plt.legend()
plt.show()


# ===============================
# Collect all model results
# Replace the numbers with your actual results
# ===============================
model_results = {
    "Model": ["XGBoost", "Random Forest", "CatBoost", "Logistic Regression", "SVM"],
    "Accuracy": [0.8142, 0.7723, 0.7596, 0.6440, 0.6831],
    "AUC": [0.6877, 0.6976, 0.6725, 0.6739, 0.6425],
    "Precision (Class 0)": [0.88, 0.88, 0.88, 0.91, 0.88],
    "Precision (Class 1)": [0.32, 0.29, 0.27, 0.25, 0.23],
    "Recall (Class 0)": [0.85, 0.85, 0.83, 0.64, 0.72],
    "Recall (Class 1)": [0.35, 0.35, 0.35, 0.65, 0.46],
    "F1-Score (Class 0)": [0.86, 0.86, 0.85, 0.75, 0.79],
    "F1-Score (Class 1)": [0.32, 0.32, 0.31, 0.36, 0.31]
}

# Convert to DataFrame
results_table = pd.DataFrame(model_results)

import matplotlib.pyplot as plt
import seaborn as sns

# Set style for plots
sns.set_style("whitegrid")

plt.figure(figsize=(14, 6))

# Plot 1: Accuracy Comparison
plt.subplot(1, 2, 1) # 1 row, 2 columns, 1st plot
sns.barplot(x='Model', y='Accuracy', data=results_table, palette='viridis')
plt.title('Model Accuracy Comparison')
plt.ylabel('Accuracy')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 1) # Accuracy is between 0 and 1

# Plot 2: AUC Comparison
plt.subplot(1, 2, 2) # 1 row, 2 columns, 2nd plot
sns.barplot(x='Model', y='AUC', data=results_table, palette='magma')
plt.title('Model AUC Comparison')
plt.ylabel('AUC')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 1) # AUC is between 0 and 1

plt.tight_layout() # Adjust layout to prevent overlapping titles/labels
plt.show()


# Display table
print("=== All Models Comparison ===")
display(results_table)
#DL AND ML

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import Input, BatchNormalization

dl_model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(1, activation='sigmoid')
])

dl_model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

history = dl_model.fit(
    X_train_scaled, y_train_res,
    epochs=50,
    batch_size=32,
    validation_split=0.2,
    verbose=0
)

y_pred_dl = (dl_model.predict(X_test_scaled) > 0.5).astype(int)

print(confusion_matrix(y_test, y_pred_dl))
print(classification_report(y_test, y_pred_dl))

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Input
from tensorflow.keras.optimizers import Adam
from sklearn.utils import class_weight
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_curve
import numpy as np

# ===============================
# Apply SMOTE to DL data (already done for ML)
# ===============================
# X_train_res, y_train_res already SMOTE-applied in your previous code

# ===============================
# Scale data
# ===============================
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_res)
X_test_scaled = scaler.transform(X_test)

# ===============================
# Compute class weights
# ===============================
class_weights = class_weight.compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_train_res),
    y=y_train_res
)
class_weights = dict(enumerate(class_weights))

# ===============================
# DL Model with Input layer
# ===============================
dl_model = Sequential([
    Input(shape=(X_train_scaled.shape[1],)),  # Correct way to define input
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(1, activation='sigmoid')
])

dl_model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# ===============================
# Train DL model with class weights
# ===============================
history = dl_model.fit(
    X_train_scaled,
    y_train_res,
    epochs=50,
    batch_size=32,
    validation_split=0.2,
    class_weight=class_weights,
    verbose=1
)

# ===============================
# Predictions
# ===============================
y_pred_proba_dl = dl_model.predict(X_test_scaled).flatten()

# Optimize threshold based on F1-score
precisions, recalls, thresholds = precision_recall_curve(y_test, y_pred_proba_dl)
f1_scores = 2 * precisions * recalls / (precisions + recalls + 1e-8)
best_thresh = thresholds[np.argmax(f1_scores)]
y_pred_dl = (y_pred_proba_dl >= best_thresh).astype(int)

# ===============================
# Metrics
# ===============================
print("\n=== DL Model Results After Improvement ===")
print("Best Threshold for Class 1:", best_thresh)
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_dl))
print("\nClassification Report:\n", classification_report(y_test, y_pred_dl))


from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam

dnn_model = Sequential([
    Dense(128, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    BatchNormalization(),
    Dropout(0.3),
    Dense(64, activation='relu'),
    BatchNormalization(),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

dnn_model.compile(optimizer=Adam(learning_rate=0.001),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])



# ============================================================
# 14. FEATURE SELECTION (SAFE)
# ============================================================

from sklearn.ensemble import RandomForestClassifier

# Retrain RF on the current feature set (X_train_res)
rf_model_fs = RandomForestClassifier(
    n_estimators=300,
    max_depth=8,
    random_state=42,
    class_weight='balanced'
)
rf_model_fs.fit(X_train_res, y_train_res)

# Get feature importance
feat_imp_df = pd.DataFrame({
    "Feature": X_train_res.columns,
    "Importance": rf_model_fs.feature_importances_
}).sort_values(by="Importance", ascending=False)

# Select top 10 features
top_features = feat_imp_df.head(10)["Feature"].tolist()
print("Top 10 Selected Features:", top_features)

X_train_fs = X_train_res[top_features].copy()
X_test_fs = X_test[top_features].copy()

# ============================================================
# 15. NOISE REMOVAL (Z-SCORE METHOD)
# ============================================================

from scipy.stats import zscore

z_scores = np.abs(zscore(X_train_fs))
mask = (z_scores < 3).all(axis=1)

X_train_clean = X_train_fs.loc[mask]
y_train_clean = y_train_res.loc[mask]

print("Shape after noise removal:", X_train_clean.shape)

# ============================================================
# 16. DATA AUGMENTATION (FOR DL)
# ============================================================

import numpy as np

noise_factor = 0.01
X_train_augmented = X_train_scaled + noise_factor * np.random.normal(
    0, 1, X_train_scaled.shape
)
print("DL Data Augmentation applied (Gaussian noise)")

# ============================================================
# 17. TRAINING TIME COMPARISON
# ============================================================

import time

training_times = {}

# XGBoost
start = time.time()
xgb_model.fit(X_train_res, y_train_res)
training_times["XGBoost"] = time.time() - start

# Random Forest
start = time.time()
rf_model_fs.fit(X_train_res, y_train_res)
training_times["Random Forest"] = time.time() - start

# CatBoost
start = time.time()
cat_model.fit(X_train_res, y_train_res, cat_features=categorical_cols, verbose=0)
training_times["CatBoost"] = time.time() - start

# Logistic Regression
start = time.time()
best_log_model.fit(X_train_res, y_train_res)
training_times["Logistic Regression"] = time.time() - start

# SVM
start = time.time()
svm_model.fit(X_train_scaled, y_train_res)
training_times["SVM"] = time.time() - start

# Deep Learning
start = time.time()
dl_model.fit(
    X_train_scaled,
    y_train_res,
    epochs=10,  # reduced for time measurement
    batch_size=32,
    verbose=0
)
training_times["Deep Learning"] = time.time() - start

print("\nTraining Time (seconds):")
for k, v in training_times.items():
    print(f"{k}: {v:.2f}")

# ============================================================
# 18. MODEL COMPLEXITY
# ============================================================

model_complexity = {
    "XGBoost": f"{xgb_model.n_estimators} trees",
    "Random Forest": f"{rf_model_fs.n_estimators} trees",
    "CatBoost": f"{cat_model.tree_count_} trees",
    "Logistic Regression": "Low (linear)",
    "SVM": "Depends on support vectors",
    "Deep Learning": f"{dl_model.count_params()} parameters"
}

print("\nModel Complexity:")
for k, v in model_complexity.items():
    print(f"{k}: {v}")

# ============================================================
# 19. FINAL COMPARISON TABLE
# ============================================================

final_results = pd.DataFrame({
    "Model": list(training_times.keys()),
    "Training Time (s)": list(training_times.values()),
    "Model Complexity": list(model_complexity.values())
})

display(final_results)

# ============================================================
# 20. TRAINING TIME VISUALIZATION
# ============================================================

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
sns.barplot(
    x="Model",
    y="Training Time (s)",
    data=final_results
)
plt.xticks(rotation=45)
plt.title("Training Time Comparison")
plt.show()





import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ===============================
# Training Time and Complexity Table
# ===============================
final_results = pd.DataFrame({
    "Model": ["XGBoost", "Random Forest", "CatBoost", "Logistic Regression", "SVM", "Deep Learning"],
    "Training Time (s)": [1.48, 3.18, 2.66, 3.91, 5.48, 4.40],
    "Model Complexity": ["200 trees", "300 trees", "500 trees", "Low (linear)", "Depends on support vectors", "3329 parameters"]
})

# Display the table
display(final_results)

# ===============================
# Training Time Visualization
# ===============================
plt.figure(figsize=(10,5))
sns.barplot(x="Model", y="Training Time (s)", data=final_results, palette="viridis")
plt.title("Training Time Comparison Across Models")
plt.ylabel("Training Time (seconds)")
plt.xticks(rotation=45)
plt.ylim(0, 6)
plt.show()
```
